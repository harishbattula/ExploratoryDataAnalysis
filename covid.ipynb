```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("COVID-19 Coronavirus.csv", usecols = ['Country', 'ISO 3166-1 alpha-3 CODE', 'Population',
           'Continent', 'Total Cases', 'Total Deaths','Death percentage'])
df.head()
```

```python
df.columns
```

```python
for column in df.columns:
    print("\n"+column)
    print(df[column].value_counts())
```

```python
df.describe()
```

```python
# Data is already in clear manner
# We do some feature engineering stuff

# Round Death percentage  values to 2
df["Death percentage"] = df["Death percentage"].apply(lambda x: round(x, 2))
df["Death percentage"]
```

```python
# It makes sense to add alive percentage feature to our data
df["Alive percentage"] = 100 - df["Death percentage"] 
df["Alive percentage"]
```

```python
df["Affected Percentage"] = round(((df["Total Cases"] / df["Population"]) * 100), 2)
df["Affected Percentage"]
```

```python
df.columns
```

```python
# There are 34 countries with nick name differ from actual name
# so we cant drop Other name feature

# For simplicity change ISO 3166-1 alpha-3 CODE to country code
df["Country Code"] = df["ISO 3166-1 alpha-3 CODE"]
df.drop(columns = ["ISO 3166-1 alpha-3 CODE"],axis = 1, inplace=True)
df.columns
```

```python
df["Unaffected Parcentage"] = 100 - df["Affected Percentage"]
df["Unaffected Parcentage"]
```

```python
# sns.set_theme(style="whitegrid")
# tips = sns.load_dataset("tips")
# ax = sns.barplot(x="day", y="total_bill", data=tips)

XValues = df["Continent"].value_counts()
YValues = list()

for continent in XValues.index :
    YValues.append(df[df["Continent"]==continent]["Total Cases"].sum())
```

```python
sns.set_theme(style = "whitegrid")
barplot = sns.barplot(x = XValues.index, y = YValues)
barplot.set_title("Continent basis Cases")
barplot.set_xlabel("Continent")
barplot.set_ylabel("No of Cases")
barplot.set_xticklabels(barplot.get_xticklabels(), rotation=90)
plt.plot()
```

```python
print("Total No of cases in Asia Continent ",df[df["Continent"] == "Asia"]["Total Cases"].sum())
print("Total No of cases in Europe Continent ",df[df["Continent"] == "Europe"]["Total Cases"].sum())
```

```python
# Total no of deaths with respect to continent
X = df["Continent"].value_counts()
Y = list()
for continent in X.index:
    Y.append(df[df["Continent"] == continent]["Total Deaths"].sum())

bar = sns.barplot(x = X.index, y = Y)
bar.set_title("Total Deaths with respect to each continent")
bar.set_xlabel("Continent")
bar.set_ylabel("Total Deaths")
bar.set_xticklabels(bar.get_xticklabels(), rotation = 90)
bar.plot()
```

```python
df.columns
```

```python
# From the above plot it is clear that we don't have deaths in oceania
# lets plot unaffected people percentage across all continents
UnAffectedPercent = list()
for continent in X.index:
    UnAffectedPercent.append(df[df["Continent"] == continent]["Unaffected Parcentage"].mean())

bar = sns.barplot(x = X.index, y = UnAffectedPercent)
bar.set_title("Unaffected Percentage across each continent")
bar.set_xlabel("Continent")
bar.set_ylabel("Percentage")
bar.set_xticklabels(bar.get_xticklabels(), rotation = 90)
bar.plot()
```

```python
# from the above plot, we can say that africa people are not much affected by covid
# and europe people are more affected by covid

print("Displaying First barplot in using pivot tabels")
pd. set_option('display.float_format', lambda x: '%.2f' % x)
pd.pivot_table(df, index = 'Continent', values = 'Total Cases', aggfunc="sum")
```

```python
# we can use as many as columns until we have sensable meaning to our pivot table
pd.set_option("display.max_rows",250)
pd.pivot_table(df, index = ["Continent","Country Code"], values = "Total Cases", aggfunc="sum")
```
